```python
from matplotlib import pyplot as plt
import numpy as np

n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X)

plt.plot(X, Y + 1, color='cyan', alpha=1.00)
plt.fill_between(X, Y + 1, 1, color='cyan', alpha=0.2)
plt.plot(X, Y - 1, color='blue', alpha=1.00)
plt.fill_between(X, Y -1, -1, color='blue', alpha=0.2, where=Y-1 >= -1)
plt.fill_between(X, Y -1, -1, color='red', alpha=0.2, where=Y-1 <= -1)

plt.xlim(-np.pi, np.pi)
plt.xticks(())
plt.ylim(-2.5, 2.5)
plt.yticks(())

plt.show()
```
