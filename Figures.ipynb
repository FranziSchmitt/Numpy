```python
from matplotlib import pyplot as plt
import numpy as np

n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X)

plt.plot(X, Y + 1, color='cyan', alpha=1.00)
plt.fill_between(X, Y + 1, 1, color='cyan', alpha=0.2)
plt.plot(X, Y - 1, color='blue', alpha=1.00)
plt.fill_between(X, Y -1, -1, color='blue', alpha=0.2, where=Y-1 >= -1)
plt.fill_between(X, Y -1, -1, color='red', alpha=0.2, where=Y-1 <= -1)

plt.xlim(-np.pi, np.pi)
plt.xticks(())
plt.ylim(-2.5, 2.5)
plt.yticks(())

plt.show()
```

```python
import matplotlib as m


n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)


t = np.arctan2(Y, X)

plt.scatter(X,Y, cmap='jet', c=t, alpha=0.2, s=10)

plt.show()
```

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X, Y1):
    plt.text(x + 0.0, y + 0.05, '%.2f' % y, ha='center', va='bottom')

for x, y in zip(X, Y2):
    plt.text(x + 0.0, -y - 0.05, '%.2f' % y, ha='center', va='top')
    
plt.ylim(-1.25, +1.25)
```

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)

plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap='hot')
C = plt.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.5)
plt.clabel(C, fontsize=7)
```
